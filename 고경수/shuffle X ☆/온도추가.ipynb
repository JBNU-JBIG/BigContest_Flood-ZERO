```python
import pandas as pd
import numpy as np

from catboost import CatBoostRegressor
cat = CatBoostRegressor()

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('G:/내 드라이브/Github/BigContest_Flood-ZERO/고경수/train.csv')
test = pd.read_csv('G:/내 드라이브/Github/BigContest_Flood-ZERO/고경수/test.csv')
submission = pd.read_csv('G:/내 드라이브/Github/BigContest_Flood-ZERO/고경수/submission.csv')

temp = pd.read_csv('G:/내 드라이브/Github/BigContest_Flood-ZERO/전국기온.csv')
```

```python
train['연'] = train['연'].astype(float)
train['월'] = train['월'].astype(float)
train['일'] = train['일'].astype(float)

train['연'] = train['연'].astype(int)
train['월'] = train['월'].astype(int)
train['일'] = train['일'].astype(int)

train['홍수사상번호'] = train['홍수사상번호'].astype(str)
train['연'] = train['연'].astype(str)
train['월'] = train['월'].astype(str)
train['일'] = train['일'].astype(str)
train['시간'] = train['시간'].astype(str)

test['연'] = test['연'].astype(str)
test['홍수사상번호'] = test['홍수사상번호'].astype(str)
test['월'] = test['월'].astype(str)
test['일'] = test['일'].astype(str)
test['시간'] = test['시간'].astype(str)
```

```python
train['일자'] = train['연']+'-'+train['월']+'-'+train['일']
```

```python
train['일자'] = pd.to_datetime(train['일자'])
```

```python
temp.columns = ['일자','평균기온']
```

```python
temp['일자'] = pd.to_datetime(temp['일자'])
```

```python
train['평균기온'] = 0.0
```

```python
temp.loc[temp['일자'] == train['일자'][50], '평균기온']
```

```python
for i in range(len(train)):
    train['평균기온'][i] = temp.loc[temp['일자'] == train['일자'][i], '평균기온']
```

```python
train['정규평균기온'] = (train['평균기온'] - train['평균기온'].min()) / (train['평균기온'].max()-train['평균기온'].min())
```

```python
유역평균강수 = ['유역평균강수1','유역평균강수2','유역평균강수3','유역평균강수4','유역평균강수5','유역평균강수6']
강우A = ['강우A1','강우A2','강우A3','강우A4','강우A5','강우A6']
강우B = ['강우B1','강우B2','강우B3','강우B4','강우B5','강우B6']
강우C = ['강우C1','강우C2','강우C3','강우C4','강우C5','강우C6']
강우D = ['강우D1','강우D2','강우D3','강우D4','강우D5','강우D6']
수위E = ['수위E1','수위E2','수위E3','수위E4','수위E5','수위E6']
수위D = ['수위D1','수위D2','수위D3','수위D4','수위D5','수위D6']
```

```python
X_train, X_test, y_train, y_test = train_test_split(train[['정규평균기온','평균기온','홍수사상번호','연','월','일','시간',*유역평균강수, *강우A, *강우B, *강우C, *강우D, *수위E, *수위D]],
                                                           train['유입량'], test_size=0.2109996, shuffle=False)

Original_train = X_train # 사상번호 1 ~ 21
Original_test = X_test # 사상번호 22 ~ 25
```

---
# 온도 추가 집단5

```python
X_train = Original_train[['정규평균기온', 유역평균강수[4], 강우A[4], 강우B[4], 강우C[4], 강우D[4], 수위E[4], 수위D[4]]]
X_test = Original_test[['정규평균기온', 유역평균강수[4], 강우A[4], 강우B[4], 강우C[4], 강우D[4], 수위E[4], 수위D[4]]]

categorical_features_indices1 = np.where(X_train.dtypes == np.object)[0]

model = cat
model.fit(X_train, y_train, cat_features = categorical_features_indices1)
```

```python
pred = model.predict(X_test)
```

```python
rmse = (np.sqrt(np.mean(mean_squared_error(y_test, pred))))
rmse
```
